```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#load image
image = cv2.imread("D:\\IS Project\\Quality-monitoring-of-mangoes-using-normal-images\\Mango Image data\\Plasma Treated Image data\\2021-08-04 13_11_45.jpg")
image_ = image[0:1808,0:2188]
# Display the image
plt.imshow(image_)
# Show the plot
plt.show()
#convert image to HSV
HSV = cv2.cvtColor(image_,cv2.COLOR_BGR2HSV)
H,S,V = cv2.split(HSV)
mean_h,std_h = cv2.meanStdDev(H)
mean_S,std_S = cv2.meanStdDev(S)
mean_v,std_v = cv2.meanStdDev(V)
#ranges
Upper_h = mean_h[0][0]+std_h[0][0]
Lower_h = mean_h[0][0]-1.5*std_h[0][0]
Upper_s = mean_S[0][0]+3*std_S[0][0]
Lower_s = mean_S[0][0]-2.1*std_S[0][0]
Upper_v = mean_v[0][0]+3*std_v[0][0]
Lower_v = mean_v[0][0]-std_v[0][0]
mask = cv2.inRange(HSV,(Lower_h,Lower_s,Lower_v),(Upper_h,Upper_s,Upper_v))

# Display the image
plt.imshow(mask)
# Show the plot
plt.show()

result = cv2.bitwise_and(image_,image_,mask=mask)
# Display the image
plt.imshow(result)
# Show the plot
plt.show()
```
